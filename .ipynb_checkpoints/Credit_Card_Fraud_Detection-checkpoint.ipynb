```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.preprocessing import StandardScaler
from sklearn.cluster import DBSCAN
from sklearn.ensemble import IsolationForest
from sklearn.neighbors import NearestNeighbors
from scipy.spatial import distance
from datetime import datetime
from mpl_toolkits.mplot3d import Axes3D
import os
import random
import warnings
warnings.filterwarnings('ignore')
random.seed(42)
seed = 42
```

```python
# III. Data precprocessing
def form_scaled_df(card, time_blocks=200, scaler=StandardScaler()):
    q_diff = 1/time_blocks
    q = list(np.arange(0, 1+q_diff, q_diff))
    labels = range(len(q)-1)
    card['time_block'] = pd.qcut(card['Time'], q=q, labels=labels)
    df = card.drop(columns=['Class', 'Time', 'time_block'])
    columns = df.columns.tolist()
    scaled_columns = ['scaled_' + col for col in columns]
    scaled_df = scaler.fit_transform(df[columns].values)
    scaled_df = np.column_stack((scaled_df, card[['Class', 'Time', 'time_block']].values))
    scaled_df = pd.DataFrame(scaled_df, columns=scaled_columns + ['Class', 'Time', 'time_block'])
    return scaled_df, scaled_columns, labels
scaled_df, scaled_columns, labels = form_scaled_df(card)
```

```python
def features_selection(scaled_df, scaled_columns, threshold=492):
    target_anomalies_ix = scaled_df[scaled_df['Class'] == 1].index.tolist()
    bdict = {}
    adict = {}
    counts = []
    for col in scaled_columns:
        scaled_df[col] = abs(scaled_df[col])
        top_anomalies = scaled_df.sort_values(by=col, 
                                              ascending=False)[:threshold]
        top_anomalies_ix = top_anomalies.index.tolist()
        bdict[col] = top_anomalies_ix
        intersect_anomalies_ix = [ix for ix in top_anomalies_ix if ix in target_anomalies_ix]
        adict[col] = intersect_anomalies_ix
        count = len(intersect_anomalies_ix)
        counts.append(count)
    summary = pd.DataFrame({'feature': scaled_columns, 'cnt': counts}).sort_values(by='cnt', 
                                                                                   ascending=False)
    accum_anomalies_ix = []
    accum_ix = []
    selected_features = []
    cdict = {'feature': [], 'accum_anom': [], 'accum_obs': [], 'precision': [], 'recall': [], 'f1': []}
    for row in summary.values:
        selected_features.append(row[0])
        for ix in bdict[row[0]]:
            if ix not in accum_ix:
                accum_ix.append(ix)
        for ix in adict[row[0]]:
            if ix not in accum_anomalies_ix:
                accum_anomalies_ix.append(ix)
        accum_anom = len(accum_anomalies_ix)
        accum_obs = len(accum_ix)
        precision = round(len(accum_anomalies_ix)/len(accum_ix), 3)
        recall = round(len(accum_anomalies_ix)/threshold, 3)
        f1 = 2 * (precision * recall)/(precision + recall)
        cdict['feature'].append(row[0])
        cdict['accum_anom'].append(accum_anom)
        cdict['accum_obs'].append(accum_obs)
        cdict['precision'].append(precision)
        cdict['recall'].append(recall)
        cdict['f1'].append(f1)
        if len(accum_anomalies_ix) == len(target_anomalies_ix):
            break
    summary = pd.DataFrame(cdict)
    summary.to_csv('./output/tab1.csv', index=False)
    return summary, selected_features
summary, selected_features = features_selection(scaled_df, scaled_columns, threshold=492)
summary
```

```python
# select 3 first features (cover most anomalies as using 492 most distant observations)
scaled_columns = selected_features[:3]
scaled_columns
```

```python
# IV. Model development
# Distance-based method
def find_anomalies_using_distance_based(scaled_df, scaled_columns, top_n=492):
    print(f"Model: Distance-based method, Parameters: top_n={top_n}")
    X = scaled_df[scaled_columns].values
    center = np.mean(X, axis=0)
    dist = np.sum((X - center)**2, axis = 1)**0.5
    scaled_df['dist'] = dist
    scaled_df = scaled_df.sort_values(by='dist', ascending=False)
    anomalies_df = scaled_df[:top_n]
    anomalies_ix = anomalies_df.index.tolist()
    correct_anomalies = int(sum(anomalies_df['Class']))
    actual_anomalies = int(sum(scaled_df['Class']))
    precision = round(correct_anomalies/top_n, 3)
    recall = round(correct_anomalies/actual_anomalies, 3)
    f1 = 2 * (precision * recall)/(precision + recall)
    print(f"Correct anomalies: {correct_anomalies}, ", 
          f"Pred anomalies: {top_n}, ", 
          f"Act anomalies: {actual_anomalies}", 
          f"% Prec: {precision}, ", 
          f"% Recall: {recall}, ", 
          f"F1-Score: {f1}")
    print("---------------------------------------------------------")
    return anomalies_ix
```

```python
def calculate_fpr_tpr(scaled_df, anomalies_ix):
    normal_ix = [ix for ix in scaled_df.index.tolist() if ix not in anomalies_ix]
    tp = sum(scaled_df.iloc[anomalies_ix, :]['Class'])
    fp = len(anomalies_ix) - tp
    fn = sum(scaled_df.iloc[normal_ix, :]['Class'])
    tn = len(normal_ix) - fn
    fpr = fp/(fp + tn)
    tpr = tp/(tp + fn)
    return fpr, tpr
```

```python
def arrange_fprs_tprs(fprs, tprs):
    tab = pd.DataFrame({'fprs': fprs, 'tprs': tprs})
    tab = tab.sort_values(by='fprs', ascending=True)
    fprs, tprs = tab['fprs'].tolist(), tab['tprs'].tolist()
    if 0 not in fprs:
        fprs = [0] + fprs
    if 1 not in fprs:
        fprs += [1]
    if 0 not in tprs:
        tprs = [0] + tprs
    if 1 not in tprs:
        tprs += [1]
    return fprs, tprs
```

```python
def calculate_auc(fprs, tprs):
    auc = 0
    for i in range(len(fprs)-1):
        height = abs(fprs[i] - fprs[i+1])
        auc += (tprs[i] + tprs[i+1])*height/2
    return auc
```

```python
def calculate_distance_based_fprs_tps(scaled_df, scaled_columns, 
                              top_ns=range(200, 1200, 100)):
    fprs, tprs = [], []
    for top_n in top_ns:
        start_time = datetime.now()
        anomalies_ix = find_anomalies_using_distance_based(scaled_df, scaled_columns, top_n=top_n)
        end_time = datetime.now()
        duration = (end_time-start_time).total_seconds()
        print(f"Duration: {round(duration, 2)} (s)")
        print("---------------------------------------------------------")
        fpr, tpr = calculate_fpr_tpr(scaled_df, anomalies_ix)
        fprs.append(fpr)
        tprs.append(tpr)
    fprs, tprs = arrange_fprs_tprs(fprs, tprs)
    auc = calculate_auc(fprs, tprs)
    print(f"AUC of Distance-based method: {auc}")
    return fprs, tprs
```

```python
distance_based_fprs, distance_based_tprs = calculate_distance_based_fprs_tps(scaled_df, scaled_columns, 
                                                     top_ns=range(200, 1200, 100))
```

```python
# Isolation Forest
def find_anomalies_using_IsolationForest(scaled_df, scaled_columns, 
                                         contamination=0.017):
    print(f"Model: Isolation Forest, Parameters: max_features={len(scaled_columns)}, contamination={contamination}")
    X = scaled_df[scaled_columns].values
    forest = IsolationForest(max_features=len(scaled_columns), 
                             contamination=contamination, random_state=seed)
    forest.fit(X)
    scaled_df['cluster'] = forest.predict(X)
    anomalies_ix = scaled_df[scaled_df['cluster'] == -1].index.tolist()
    no_actual_anomalies = int(sum(scaled_df['Class']))
    no_pred_anomalies = len(anomalies_ix)
    no_correct_anomalies = int(sum(scaled_df.iloc[anomalies_ix, :]['Class']))
    precision = round(no_correct_anomalies/no_pred_anomalies, 3)
    recall = round(no_correct_anomalies/no_actual_anomalies, 3)
    f1 = 2 * (precision * recall)/(precision + recall)
    print(f"Correct anomalies: {no_correct_anomalies}, ", 
          f"Pred anomalies: {no_pred_anomalies}, ", 
          f"Act anomalies: {no_actual_anomalies}",
          f"% Prec: {precision}, ", 
          f"% Recall: {recall}, ",
          f"F1-Score: {f1}")
    print("---------------------------------------------------------")
    return anomalies_ix
```

```python
def calculate_forest_fprs_tps(scaled_df, scaled_columns, 
                              contaminations=np.arange(0.0005, 0.0055, 0.0005)):
    fprs, tprs = [], []
    for contamination in contaminations:
        contamination = round(contamination, 4)
        start_time = datetime.now()
        anomalies_ix = find_anomalies_using_IsolationForest(scaled_df, scaled_columns, 
                                                            contamination=contamination)
        end_time = datetime.now()
        duration = (end_time-start_time).total_seconds()
        print(f"Duration: {round(duration, 2)} (s)")
        print("---------------------------------------------------------")
        fpr, tpr = calculate_fpr_tpr(scaled_df, anomalies_ix)
        fprs.append(fpr)
        tprs.append(tpr)
    fprs, tprs = arrange_fprs_tprs(fprs, tprs)
    auc = calculate_auc(fprs, tprs)
    print(f"AUC of Isolation Forest: {auc}")
    return fprs, tprs
```

```python
forest_fprs, forest_tprs = calculate_forest_fprs_tps(scaled_df, scaled_columns, 
                                                     contaminations=np.arange(0.0005, 0.0055, 0.0005))
```

```python
# DBSCAN
def fit_DBSCAN(db, scaled_df, scaled_columns, ix):
    sub_df = scaled_df[scaled_columns].iloc[ix, :]
    X = sub_df.values
    db.fit(X)
    sub_df['cluster'] = db.labels_
    anomalies_ix = sub_df[sub_df['cluster'] == -1].index.tolist()
    return anomalies_ix
```

```python
def find_anomalies_using_DBSCAN(scaled_df, scaled_columns, labels,
                                eps=0.8, min_samples=10, anomalies_pool_ix=[], retries=2):
    if retries == 2:
        print(f"Model: DBSCAN, Parameters: eps={eps}, min_samples={min_samples}")
    db = DBSCAN(eps=eps, min_samples=min_samples)
    anomalies_ix = []
    for label in labels:
        ix1 = scaled_df[scaled_columns][scaled_df['time_block'] == label].index.tolist()
        ix1 += anomalies_pool_ix
        ix1 = np.unique(ix1).tolist()
        anomalies_ix += fit_DBSCAN(db, scaled_df, scaled_columns, ix=ix1)
    if retries == 1:
        anomalies_df = pd.DataFrame({'ix': anomalies_ix})
        anomalies_df = anomalies_df.groupby(by='ix', as_index=False).agg(
            cnt = ('ix', 'count')
        )
        anomalies_ix = anomalies_df[anomalies_df['cnt'] == len(labels)]['ix'].tolist()
    else:
        anomalies_ix = np.unique(anomalies_ix).tolist()
        anomalies_ix = fit_DBSCAN(db, scaled_df, scaled_columns, ix=anomalies_ix)
    no_actual_anomalies = int(sum(scaled_df['Class']))
    no_pred_anomalies = len(anomalies_ix)
    no_correct_anomalies = int(sum(scaled_df.iloc[anomalies_ix, :]['Class']))
    precision = round(no_correct_anomalies/no_pred_anomalies, 3)
    recall = round(no_correct_anomalies/no_actual_anomalies, 3)
    f1 = 2 * (precision * recall)/(precision + recall)
    if retries == 1:
        print(f"Correct anomalies: {no_correct_anomalies}, ", 
              f"Pred anomalies: {no_pred_anomalies}, ", 
              f"Act anomalies: {no_actual_anomalies}, ", 
              f"% Prec: {precision}, ", 
              f"% Recall: {recall}, ",
              f"F1-Score: {f1}")
        print("---------------------------------------------------------")
    retries -= 1
    if retries > 0:
        anomalies_ix = find_anomalies_using_DBSCAN(scaled_df, scaled_columns, labels, 
                                                   eps=eps, min_samples=min_samples, 
                                                   anomalies_pool_ix=anomalies_ix, 
                                                   retries=retries)
    return anomalies_ix
```

```python
def calculate_dbscan_fprs_tps(scaled_df, scaled_columns, labels,
                              epss=np.arange(0.2, 2.2, 0.2), min_samples=10):
    fprs, tprs = [], []
    for eps in epss:
        eps = round(eps, 1)
        start_time = datetime.now()
        anomalies_ix = find_anomalies_using_DBSCAN(scaled_df, scaled_columns, labels,
                                                   eps=eps, min_samples=min_samples)
        end_time = datetime.now()
        duration = (end_time-start_time).total_seconds()
        print(f"Duration: {round(duration, 2)} (s)")
        print("---------------------------------------------------------")
        fpr, tpr = calculate_fpr_tpr(scaled_df, anomalies_ix)
        fprs.append(fpr)
        tprs.append(tpr)
    fprs, tprs = arrange_fprs_tprs(fprs, tprs)
    auc = calculate_auc(fprs, tprs)
    print(f"AUC of DBSCAN: {auc}")
    return fprs, tprs
```

```python
dbscan_fprs, dbscan_tprs = calculate_dbscan_fprs_tps(scaled_df, scaled_columns, labels,
                                                     epss=np.arange(0.2, 2.2, 0.2), min_samples=10)
```

**SUMMARY:**

The three methods Distance-based method, Isolation Forest, and DBSCAN seem to show similar powers in terms of identifying anomalies in credit card transactions. However, the prediction by **DBSCAN** is slightly better among the three models as its **AUC (0.9276)** and **F1-Score (0.723)**, which is originated from **Precision (0.747)** and **Recall (0.701)**, are the highest.

The predictions of the three models concentrate on observations, which are far from the center of the majority area. While, most of unidentified anomalies (in purple color) gather at the center, which is likely to be ignored. Such characteristic possibly suggests for future study that other features than V12, V14, and V17 should be further investigated to see whether those unidentified anomalies are discovered in any other dimensions.

In terms of training time by individual models, the distance-based method was the fastest one, which takes under 1 second to generate the prediction. While Isolation Forest takes around 23 seconds on average for a single training. And, DBSCAN with the leverage of batch training, the average training time is around 26 seconds, which is quite close to Isolation Forest. The limitation on training time is the main challenge of Isolation Forest and DBSCAN in comparison with distance-based method. However, in other shape of dataset, in which the anomalies are not observations being most distant, other models can outperform distance-based method.
